```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```
p = sns.load_dataset('penguins')
```

```
sns.pairplot(p);
```

```
sns.pairplot(p ,hue='sex');
```

```
sns.pairplot(p ,hue='species');
```
